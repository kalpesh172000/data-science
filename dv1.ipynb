```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('Titanic.csv')
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
plt.hist(x=df['Age'], bins=30) 
plt.xlabel('Age')
plt.ylabel('Frequency')   
```

```python
sns.histplot(data=df,x='Sex',hue='Survived',multiple='stack')
```

```python
sns.violinplot(data=df,x='Sex',y='Age',hue='Survived',split=True)
```

```python
sns.histplot(data=df,x='Age',y='Sex', hue='Survived',multiple='stack')
```

```python
sns.histplot(data=df,x='Fare',hue='Survived',multiple='stack')
## *this concludes that the high fare of the ticket has a higher chance of survival*
```

```python
sns.histplot(data=df,x='Pclass',hue='Survived',multiple='stack')
## this concludes that pclass 2 had high chance of survival and pclass 1 is at second position but pclass 3 had the lowest chance of survival.
```

```python
sns.histplot(data=df,y='Sex',hue='Survived',multiple='stack')
#this concludes that females had more chances of survival
```

```python
sns.histplot(data=df,y='Age',hue='Survived',multiple='stack')
#this concludes that the children had more chances of survival
#also middle aged people had more chances of survival than the old aged people
```

```python
sns.histplot(data=df,y='Pclass',x='Fare')
#this concludes that the fare of the ticket is directly proportional to the pclass
```

```python
sns.scatterplot(data=df,x='Age',y='Fare',hue='Survived')
```

```python
sns.histplot(data=df,x='Cabin',hue='Survived',multiple='stack')
```

```python
df.boxplot(column='Age',by='Sex')
```

```python
sns.boxplot(x="Sex",y="Age",hue="Survived",data=df)
plt.show()
```
