```python
import pandas as pd
```

```python
df = pd.read_csv('student.csv')
```

```python
df.isnull().sum()
```

```python
import numpy as np
hsc_mean =df['hsc_percentage'].mean()
hsc_mean = np.round(hsc_mean)
hsc_mean
```

```python
df['hsc_percentage'] = df['hsc_percentage'].fillna(hsc_mean)
```

```python
df
```

```python
df.isnull().sum()
```

```python
fe_mean = df['fe_percentage'].mean()
fe_mean = np.round(fe_mean)
df['fe_percentage'] = df['fe_percentage'].fillna(fe_mean)
```

```python
se_mean = df['se_percentage'].mean()
se_mean = np.round(se_mean)
df['se_percentage'] = df['se_percentage'].fillna(se_mean)
```

```python
import matplotlib.pyplot as plt
plt.boxplot(df['te_percentage'])
```

```python
df = df[df['te_percentage'].between(0,100)]
```

```python
q1 = df['fe_percentage'].quantile(0.25)
q3 = df['fe_percentage'].quantile(0.75)
iqr = q3 -q1
lower_limit = q1 - 1.5*iqr
lower_limit
upper_limit = q3 + 1.5*iqr
upper_limit
df = df[df['fe_percentage'].between(lower_limit,upper_limit)]
```

```python
df.shape
```

```python
q1= df['te_percentage'].quantile(0.25)
q3= df['te_percentage'].quantile(0.75)
iqr = q3-q1
lower_limit = q1 - 1.5*iqr
upper_limit = q3 + 1.5*iqr
df = df[df['te_percentage'].between(lower_limit,upper_limit)]
```

```python
def mm_normalize(x):
    return (x-df['scc_percentage'].min()) / (df['scc_percentage'].max()-df['scc_percentage'].min())

df.loc[:, 'scc_normalized'] = df['scc_percentage'].apply(lambda x: mm_normalize(x))
df
```

```python
# from scipy import stats as st 
# import numpy as np
# z_score = st.zscore(df['be_percentage'])
# threshold = 3

# outliers = df[((z_score > 3) | (z_score < -3))]

# be_mean = np.mean(df.loc[~df.index.isin(outliers.index), 'be_percentage'])

# df.loc[outliers.index, 'be_percentage'] = be_mean
def zz_score(x,mean,std):
    return (x-mean)/std
mean = df['be_percentage'].mean()
std = df['be_percentage'].std()
df['be_zscore'] = df['be_percentage'].apply(lambda x: zz_score(x,mean,std))
outliers = df[(df['be_zscore'] > 3) | (df['be_zscore'].abs() < -3)]
be_mean = np.mean(df.loc[(df['be_zscore'] < 3) & (df['be_zscore']> -3), 'be_percentage'])
be_mean
df.loc[df.index.isin(outliers.index),'be_percentage'] = be_mean
```

```python
df['be_percentage'][0] = 131
```

```python
q1 = df['se_percentage'].quantile(0.25)
q3 = df['se_percentage'].quantile(0.75)
iqr = q3 - q1
lower_limit = q1 - 1.5*iqr
upper_limit = q3 + 1.5*iqr
outliers = df.loc[~df['se_percentage'].between(lower_limit, upper_limit)] 
se_mean = np.mean(df.loc[~df.index.isin(outliers.index), 'se_percentage'])
df.loc[outliers.index, 'se_percentage'] = se_mean
```

```python
def zscore(x,mean,std):
    return (x-mean)/std

mean = df['te_percentage'].mean()
std = df['te_percentage'].std()

df['te_normalize'] = df['te_percentage'].apply(lambda x: zscore(x,mean,std))
```
